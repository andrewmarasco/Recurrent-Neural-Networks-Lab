## Lab: RNN with Forecasting

### Wind Turbine Power Prediction with RNN

This lab uses a Recurrent Neural Network to predict wind turbine power output.
* [Data link](https://www.kaggle.com/code/ahmedfathygwely/wind-turbine-dataset-machine-learning-rnn-times/input)

### Step 0: Load libraries & Data

```python
# Step 0: Load libraries & Data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import math

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Load the data
df = pd.read_csv('T1.csv')
```

```python
# Display basic information
df.shape
df.head()
```

### Step 1: Prepare the data

```python
# Step 1: Prepare the data
# Converting to proper date/time format
df['Date/Time'] = pd.to_datetime(df['Date/Time'], format='%d %m %Y %H:%M')

# checking for and handling null/missing values
df.isnull().sum()
df = df.dropna()

# Basic EDA
df.describe()
```

```python
# Plot relationships between Wind Speed and Active Power
plt.figure()
plt.scatter(df['Wind Speed (m/s)'], df['LV ActivePower (kW)'], alpha=0.3)
plt.xlabel('Wind Speed (m/s)')
plt.ylabel('LV ActivePower (kW)')
plt.title('Wind Speed vs Active Power')
plt.show()
```

```python
# Selecting relevant columns
data = df[['LV ActivePower (kW)', 'Wind Speed (m/s)', 'Theoretical_Power_Curve (KWh)']]

# Scale to 0-1
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Back to dataframe
scaled_df = pd.DataFrame(scaled_data, columns=data.columns, index=df.index)

scaled_df.head()
```

### Step 2: Create sequences

```python
# Step 2: Create sequences

def create_sequences(data, seq_length, target_col=0):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length])
        y.append(data[i + seq_length, target_col])
    return np.array(X), np.array(y)

seq_length = 24

X, y = create_sequences(scaled_data, seq_length)

X.shape, y.shape
```

### Step 3: Split the data

```python
# Step 3: Split the data
# 80/20 train-test split

train_size = int(len(X) * 0.8)

X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Step 4: Build the RNN Model

```python
# Step 4: Build the RNN Model

model = Sequential()

# 1st LSTM layer
model.add(LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))

# 2nd LSTM layer
model.add(LSTM(32))
model.add(Dropout(0.2))

# Final Dense layer
model.add(Dense(1))

# Compiling
model.compile(optimizer='adam', loss='mean_squared_error')

# Summary of model
model.summary()
```

### Step 5: Train the model

```python
# Step 5: Train the model

# Callbacks
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

checkpoint = ModelCheckpoint(
    filepath='best_lstm_model.keras',
    monitor='val_loss',
    save_best_only=True
)

callbacks = [early_stop, checkpoint]

# Fit the model

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=callbacks,
    verbose=1
)

```

```python
plt.figure()
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

### Step 6: Evaluate on the test data and visualize the results

```python
# Step 6: Evaluate and visualize
# Generate predictions
y_pred = model.predict(X_test)
```

```python
# Inverse-scaling predictions and true values

# Ensure shapes are compatible
y_test_reshaped = y_test.reshape(-1, 1)

# Creating placeholders (3 columns)
y_test_placeholder = np.zeros((len(y_test), scaled_data.shape[1]))
y_pred_placeholder = np.zeros((len(y_test), scaled_data.shape[1]))

# Putting Values into target columns (0 = LV ActivePower)
y_test_placeholder[:, 0] = y_test
y_pred_placeholder[:, 0] = y_pred.flatten()

# Inverse transform back to original units
y_test_inv = scaler.inverse_transform(y_test_placeholder)[:, 0]
y_pred_inv = scaler.inverse_transform(y_pred_placeholder)[:, 0]
```

```python
# Metrics : MSE RMSE, R2
mse = mean_squared_error(y_test_inv, y_pred_inv)
rmse = math.sqrt(mse)
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"MSE:  {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R^2:  {r2:.4f}")
```

```python
# Plotting the Time Series (Actual & Predicted)

plt.figure()
plt.plot(y_test_inv, label="Actual")
plt.plot(y_pred_inv, label="Predicted")
plt.title("Actual vs Predicted Power Output (Test Set)")
plt.xlabel("Time Step")
plt.ylabel("Power (kW)")
plt.legend()
plt.show()
```

```python
# Plotting Scatter (Actual & Predicted)

plt.figure()
plt.scatter(y_test_inv, y_pred_inv, alpha=0.3)
plt.title("Predicted vs Actual (Test Set)")
plt.xlabel("Actual Power (kW)")
plt.ylabel("Predicted Power (kW)")
plt.show()
```
